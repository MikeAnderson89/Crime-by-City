```python
import pandas as pd
import numpy as np
import json
import requests
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
#Read CSV files
csv1 = "DC_Violent_Crime.csv"
df1 = pd.read_csv(csv1)
            
csv2 = "DC_Population.csv"
df2 = pd.read_csv(csv2)

csv3 = "DC_Med_Income.csv"
df3 = pd.read_csv(csv3)

csv4 = "DC_Perc_Poverty.csv"
df4 = pd.read_csv(csv4)

csv5 = "DC_Grad_Rate.csv"
df5 = pd.read_csv(csv5)

df1
#df2
#df3
#df4
#df5
```

```python
df1['Year'] = pd.to_numeric(df1['Year'])
df1['Violent Crime Rate'] = pd.to_numeric(df1['Violent Crime Rate'])

#Plot violent crime rate
df1.plot(kind='line', x='Year', y='Violent Crime Rate', figsize=(15,5), color='pink')

plt.xlim(2006, 2017)
plt.ylim(300, 800)
plt.title('Crime Rate per 100,000 people')
plt.xlabel('Year')
plt.ylabel('Rate (Per 100,000 People)')

```

```python
df5['Year'] = pd.to_numeric(df5['Year'])
df5['Graduation_Rate'] = pd.to_numeric(df5['Graduation_Rate'])

#Plot the graph
df5.plot(kind='line', x='Year', y='Graduation_Rate', figsize=(15,10), color='orange')
plt.title('High School Graduation Rates')
plt.xlabel('Year')
plt.ylabel('Graduation Rate')
```

```python
#Plot the Poverty Graph
df4['Year'] = pd.to_numeric(df4['Year'])
df4['Percent Below Poverty Line'] = pd.to_numeric(df4['Percent Below Poverty Line'])

#Plot the graph
df4.plot(kind='line', x='Year', y='Percent Below Poverty Line', figsize=(15,10), color='teal')
plt.title('Percentage of Poverty')
plt.xlabel('Year')
plt.ylabel('Percent Below Poverty Line')
```

```python
df3['Year'] = pd.to_numeric(df3['Year'])
df3['Median Earnings'] = pd.to_numeric(df3['Median Earnings'])

#Plot the Median Income Graph
df3.plot(kind='line', x='Year', y='Median Earnings', figsize=(15,10), color='purple')
plt.title('Median Income')
plt.xlabel('Year')
plt.ylabel('$ Amount per Year')
```
